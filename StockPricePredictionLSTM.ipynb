```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn

data = pd.read_csv('/content/AMZN.csv')

data
```

```python
data = data[['Date', 'Close']]
data
```

```python
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
device
```

```python
data['Date'] = pd.to_datetime(data['Date'])

plt.plot(data['Date'], data['Close'])
```

```python
# the code below shows the price for that particular date and the prices for past 7 days for each date
# given Close(t-1)...Close(t-7) -> X, we are going to predict Close -> Y

from copy import deepcopy as dc

def prepare_dataframe_for_lstm(df, n_steps):
    df = dc(df)

    df.set_index('Date', inplace=True)

    for i in range(1, n_steps+1):
        df[f'Close(t-{i})'] = df['Close'].shift(i)

    df.dropna(inplace=True)

    return df

lookback = 7
shifted_df = prepare_dataframe_for_lstm(data, lookback)
shifted_df
```

```python
#  convert to numpy arrays

shifted_df_as_np = shifted_df.to_numpy()

shifted_df_as_np
```

```python
# all the features above are now going be between -1 and 1

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(-1, 1))
shifted_df_as_np = scaler.fit_transform(shifted_df_as_np)

shifted_df_as_np
```

```python
X = shifted_df_as_np[:, 1:]
y = shifted_df_as_np[:, 0]

X.shape, y.shape
X
```

```python
# mirror X horizontally to get the input data in correct
X = dc(np.flip(X, axis=1))
X
```

```python
# 95% -> train and 5% -> test
split_index = int(len(X) * 0.95)

split_index
```

```python
# split into train and test
X_train = X[:split_index]
X_test = X[split_index:]

y_train = y[:split_index]
y_test = y[split_index:]

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# get an extra dimension for all X and y
X_train = X_train.reshape((-1, lookback, 1))
X_test = X_test.reshape((-1, lookback, 1))

y_train = y_train.reshape((-1, 1))
y_test = y_test.reshape((-1, 1))

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# convert these to tensors since we are dealing with pytorch
X_train = torch.tensor(X_train).float()
y_train = torch.tensor(y_train).float()
X_test = torch.tensor(X_test).float()
y_test = torch.tensor(y_test).float()

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# The train_dataset and test_dataset instances are created to handle training and testing data
from torch.utils.data import Dataset

class TimeSeriesDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, i):
        return self.X[i], self.y[i]

train_dataset = TimeSeriesDataset(X_train, y_train)
test_dataset = TimeSeriesDataset(X_test, y_test)
```

```python
train_dataset
```

```python
# Loads data in batches of size 16 from train_dataset and test_dataset and shuffles the data for train_dataset
from torch.utils.data import DataLoader

batch_size = 16

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
# not necessary, just for visualization
for _, batch in enumerate(train_loader):
    x_batch, y_batch = batch[0].to(device), batch[1].to(device)
    print(x_batch.shape, y_batch.shape)
    break
```

```python
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_stacked_layers):
        super().__init__()
        self.hidden_size = hidden_size
        self.num_stacked_layers = num_stacked_layers

        self.lstm = nn.LSTM(input_size, hidden_size, num_stacked_layers, batch_first=True)

        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        batch_size = x.size(0)
        h0 = torch.zeros(self.num_stacked_layers, batch_size, self.hidden_size).to(device)
        c0 = torch.zeros(self.num_stacked_layers, batch_size, self.hidden_size).to(device)

        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out

model = LSTM(1, 4, 1)
model.to(device)
model
```

```python
def train_one_epoch():
    model.train(True)
    print(f'Epoch: {epoch + 1}')
    running_loss = 0.0

    for batch_index, batch in enumerate(train_loader):
        x_batch, y_batch = batch[0].to(device), batch[1].to(device)

        output = model(x_batch)
        loss = loss_function(output, y_batch)
        running_loss += loss.item()

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch_index % 100 == 99:  # print every 100 batches
            avg_loss_across_batches = running_loss / 100
            print('Batch {0}, Loss: {1:.3f}'.format(batch_index+1,
                                                    avg_loss_across_batches))
            running_loss = 0.0
    print()
```

```python
def validate_one_epoch():
    model.train(False)
    running_loss = 0.0

    for batch_index, batch in enumerate(test_loader):
        x_batch, y_batch = batch[0].to(device), batch[1].to(device)

        with torch.no_grad():
            output = model(x_batch)
            loss = loss_function(output, y_batch)
            running_loss += loss.item()

    avg_loss_across_batches = running_loss / len(test_loader)

    print('Val Loss: {0:.3f}'.format(avg_loss_across_batches))
    print('***************************************************')
    print()
```

```python
learning_rate = 0.001
num_epochs = 10
loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

for epoch in range(num_epochs):
    train_one_epoch()
    validate_one_epoch()
```

```python
import torch
import torch.nn.functional as F

def evaluate_model(model, data_loader):
    model.eval()  # Set model to evaluation mode
    total_mse = 0.0
    total_samples = 0

    with torch.no_grad():  # Disable gradient calculation for evaluation
        for inputs, targets in data_loader:
            inputs, targets = inputs.to(device), targets.to(device)

            # Forward pass: Get predictions
            outputs = model(inputs)

            # Calculate MSE
            mse = F.mse_loss(outputs, targets, reduction='sum')  # Sum the squared errors
            total_mse += mse.item()
            total_samples += len(targets)

    # Calculate average MSE
    avg_mse = total_mse / total_samples

    # Calculate RMSE
    rmse = torch.sqrt(torch.tensor(avg_mse))

    print(f'Mean Squared Error (MSE): {avg_mse:.4f}')
    print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
```

```python
def evaluate_accuracy(model, data_loader, tolerance=0.1):
    model.eval()  # Set model to evaluation mode
    correct = 0
    total = 0

    with torch.no_grad():  # Disable gradient calculation for evaluation
        for inputs, targets in data_loader:
            inputs, targets = inputs.to(device), targets.to(device)

            # Forward pass: Get predictions
            outputs = model(inputs).squeeze()  # Ensure outputs and targets are the same shape

            # Calculate the percentage error, preventing division by zero
            percentage_error = torch.abs((outputs - targets) / torch.clamp(targets, min=1e-8))

            # Check for correct predictions within tolerance (this gives a boolean tensor)
            correct_predictions = (percentage_error < tolerance)

            # Sum the number of correct predictions in this batch
            correct_in_batch = correct_predictions.sum().item()

            # Debug: Print out intermediate values for inspection
            print(f'Batch size: {targets.size(0)}, Correct in this batch: {correct_in_batch}')

            # Update the total correct and total number of samples
            correct += correct_in_batch
            total += targets.size(0)  # Total number of actual samples in this batch

    # Compute accuracy as a percentage
    if total > 0:
        accuracy = (correct / total) * 100
    else:
        accuracy = 0.0  # Handle case where no samples are present

    # Debug: Print total values for inspection
    # print(f'Total samples: {total}, Total correct predictions: {correct}')

    # print(f'Accuracy within {tolerance*100:.1f}% tolerance: {accuracy:.2f}%')
```

```python
evaluate_model(model, test_loader)
evaluate_accuracy(model, test_loader, tolerance=0.1)  # 10% tolerance
```
